```python
#IMPORTING ALL THE NECESSARY MODULES
import pandas as pd
from sklearn.model_selection import train_test_split
# label encoder
from sklearn.preprocessing import LabelEncoder
# logistics regression
from sklearn.linear_model import LogisticRegression
#joblib
import joblib
```

```python
#to read documentation of the file
with open('iris.names',mode='r') as txt:
  text_data=txt.read()
print(text_data)
```

```python
#loading the data set in pandas dataframe
iris= pd.read_csv('iris.data', names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','Target'])
iris
```

```python
#searching for the unique target values
sorted(iris['Target'].unique())
```

```python
#defining our own labelEncoder
class LabelEncoder(object):
  def __init__(self):
    self.classes_ =None
  def fit(self, labels:list)->None:
    self.classes_ =sorted(set(labels))
  def transform(self,labels)->list:
    return [self.classes_.index(label) for label in labels]
  def fit_transform(self,labels)->list:
    self.fit(labels)
    return self.transform(labels)
```

```python
iris['Target'].apply(lambda value: list(iris.Target).index(value))
```

```python
#transforming and fitting the values of target in label encoder
labeler=LabelEncoder()
iris['Target']=labeler.fit_transform(iris['Target'])
iris['Target']
```

```python
#Getting the correlaton of the target with the features
iris.corr()['Target'][:-1]
```

```python
#defining the model
lreg=LogisticRegression(max_iter=500)
model=LogisticRegression(max_iter=500)
```

```python
# make split (dataframe and series) (can also use numpy array)
Xtrain,Xtest,Ytrain,Ytest=train_test_split(iris.loc[:,iris.columns[:-1]],iris['Target'],test_size=0.33)
```

```python
Xtrain.shape,Xtest.shape,Ytrain.shape,Ytest.shape
```

```python
#Training the model and validating its score oon training and testing data
lreg.fit(Xtrain,Ytrain)
print('R2 Score (training): ',lreg.score(Xtrain,Ytrain))
print('R2 Score (testing): ',lreg.score(Xtest,Ytest))

# make model train (fit)
model.fit(Xtrain[['petal width (cm)','petal length (cm)','sepal length (cm)']],Ytrain)
print('R2 Score (training): ',model.score(Xtrain[['petal width (cm)','petal length (cm)','sepal length (cm)']],Ytrain))
print('R2 Score (testing): ',model.score(Xtest[['petal width (cm)','petal length (cm)','sepal length (cm)']],Ytest))
```

```python
#as we can see there is no difference between the accuracy of these two models therefore we can use any of them
tab= pd.DataFrame({'Predicted':lreg.predict(Xtest),'Original':Ytest.values})
tab  #showing predicted values VS Original values on test data
```

```python
filename= 'Model.sav'
joblib.dump(lreg, filename)

# Later on you can use it by the following commands

# load the model from disk
saved_model = joblib.load(filename)
result = saved_model.score(iris.loc[:,iris.columns[:-1]],iris['Target'])  #checking score on the whole dataset
print(result)
```
